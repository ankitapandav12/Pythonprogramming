```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([20,25,35,40,45,50])
y=np.array([115,130,145,150,165,180])
plt.plot(x,y)
plt.grid()
plt.title("Data Science",style="italic",size="20")
plt.show()
```

```python
x=np.array([20,25,35,40,45,50])
y=np.array([115,130,145,150,165,180])
plt.plot(x,y,marker="o",mfc="y",mec="r",linewidth="3.8",ms="11")
plt.grid(ls="--",linewidth="2",color="grey")
plt.title("Data Science",size="22",color="brown",style="italic")
plt.xlabel("x-axis",size="15")
plt.ylabel("y-axis",size="15")
plt.show()
```
