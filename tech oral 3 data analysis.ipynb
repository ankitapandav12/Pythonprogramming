# Q.1.Find the dealers data those are in resaling mode and RERA approved

```python
data=dataset.loc[(dataset["POSTED_BY"]=="Dealer") & (dataset["RERA"]==1) & (dataset["RESALE"]==1)]
data
```

```python
data["ADDRESS"]=data["ADDRESS"].str.split(",").str.get(1)
```

# Q.4.Slice the pune city and increase the target price by 5 lakh and find the value counts of bhk or rk column

```python
ad=data.loc[(data["ADDRESS"]=="Pune")]
```

```python
ad
```

```python
col1=[]
for x in ad["TARGET(PRICE_IN_LACS)"]:
    x1=x+5
    col1.append(x1)
ad.insert(12,"Price1",col1)
```

```python
ad
```

```python
ad["BHK_OR_RK"].value_counts()
```
